```python
import lightgbm as lgb
import pandas as pd
```

```python
data = pd.read_csv('./data/data_pre.csv')
```

```python
test = pd.read_csv('./data/test_pre.csv')
```

```python
def_cols = ['OrderID', 'Interval', 'ChannelID', 'Cluster', 'DeliveryType', 'prepay', 'count_edit', 'OrderCnt', 'OrderDate_weekday', 'Date_weekday']
```

```python
cols = data.columns[data.columns.isin(test.columns)]
```

```python
test = test[def_cols]
```

```python
bst = lgb.Booster(model_file='model.txt')
```

```python
pred = bst.predict(test.drop('OrderID', axis=1))
```

```python
test['Score'] = pred
```

```python
#df['Score'] = df['Score'].apply(lambda x: round(x, 2))
```

```python
test['ID'] = test['OrderID'].apply(lambda x: str(x)+' ')
```

```python
test[['ID', 'Score']].to_csv('res_.csv', index=False, encoding='windows-1251')
```
